```python
# Renaming few columns
books_df.rename(columns={'Book-Title':'title','Book-Author':'author', 'Year-Of-Publication':'year', 'Publisher':'publisher'},inplace=True)
```

```python
# Reading ratings dataset
ratings_df = pd.read_csv('/content/Ratings.csv')
```

```python
ratings_df.head()
```

```python
# Renaming few columns
ratings_df.rename(columns={'User-ID':'user_id','Book-Rating':'rating'}, inplace = True)
```

```python
ratings_df.head()
```

```python
# Checking shape of the datasets
books_df.shape
```

```python
users_df.shape
```

```python
x=ratings_df['user_id'].value_counts()>200
```

```python
x[x].shape
```

```python
y=x[x].index
```

```python
y
```

```python
ratings_df= ratings_df[ratings_df['user_id'].isin(y)]
```

```python
ratings_df.shape
```

```python
# Merging ratings & books dataset with ISBN column
ratings_with_books=ratings_df.merge(books_df,on='ISBN')
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
# Renaming a column
number_rating.rename(columns={'rating':'number of ratings'}, inplace=True)
number_rating.head()
```

```python
# Merging
final_rating=ratings_with_books.merge(number_rating,on='title')
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['number of ratings']>50]
```

```python
final_rating.shape
```

```python
# Dropping duplicates
final_rating.drop_duplicates(['user_id','title'], inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
#Replacing na cells into 0
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
book_df1 = csr_matrix(book_pivot)
```

```python
type(book_df1)
```

## **Building Book Recommendation System**

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_df1)
```

```python
book_pivot.iloc[27,:].values.reshape(1,-1)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[54,:].values.reshape(1,-1))
```

```python
distances
```

```python
suggestions ## recommendation
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
def reco(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distances,suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1))
    
    for i in range(len(suggestions)):
        if i==0:
            print("the suggestions are ",book_name,"are : ")
        if not i:
            print(book_pivot.index[suggestions[i]])
```

```python
reco('The Chamber')
```
